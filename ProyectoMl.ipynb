```python
import ast
import pandas as pd
```

```python
rows = []
with open('steam_games.json') as f:
  for line in f.readlines():
    rows.append(ast.literal_eval(line))

df = pd.DataFrame(rows)
df.head()
```

```python
generos_ano = ['genres', 'release_date']
df_genero = df[generos_ano].copy()
df_genero.fillna('', inplace=True)
```
